```python
sns.histplot(x='Number_of_Deaths' , data=data, )
plt.title('Number_of_Deaths_in_Accident')
plt.ylabel('Count')
plt.xlabel('Number_of_Deaths')
plt.show()
```

```python
sns.countplot(x='Alcohol_Involved', data=data,palette='deep')
plt.title('Alcohol_Involved or not')
plt.ylabel('number_of_Accident')
plt.xlabel('Alcohol_Involved')
plt.show()
```

```python
rc_counts = data['Road_Conditions'].value_counts()

# Plot
plt.figure(figsize=(6, 6))  # ðŸ”¹ Figure size adjust kiya
plt.pie(rc_counts, labels=rc_counts.index, autopct='%1.1f%%', 
        colors=plt.cm.Paired.colors, startangle=90)

# Title change kiya
plt.title('Road_Conditions', fontsize=14, fontweight='bold')

# Show plot
plt.show()
```

```python
values = data["Reason"].value_counts().iloc[:10]
plt.figure(figsize=(10, 6))
sns.barplot(x=values.index, y=values.values, hue=values.index, dodge=False, palette="viridis", legend=False)
plt.xlabel("Reason")
plt.ylabel("number_of_Accident")
plt.title("Top Most Common Reason")
plt.xticks(rotation=45)
plt.show()
```

```python
values = data["State"].value_counts().iloc[:10]
plt.figure(figsize=(10, 6))
sns.barplot(x=values.index, y=values.values, hue=values.index, dodge=False, palette="cool", legend=False)
plt.xlabel("State")
plt.ylabel("number_of_Accident")
plt.title("Top 10 Most Common State")
plt.xticks(rotation=45)
plt.show()
```
